```python
def most_successful(df, sport):
    # Keep only rows where a medal was awarded
    temp_df = df.dropna(subset=['Medal'])

    # Filter by sport if needed
    if sport != 'Overall':
        temp_df = temp_df[temp_df['Sport'] == sport]

    # Group by athlete name and count medals
    top_athletes = temp_df.groupby('Name').size().reset_index(name='Medal Count')

    # Get top 15 athletes
    top_athletes = top_athletes.sort_values('Medal Count', ascending=False).head(15)

    # Merge with original dataframe to get additional info like region, sport, etc.
    merged = top_athletes.merge(df, on='Name', how='left')[
        ['Name', 'Sport', 'region', 'Medal Count']
    ].drop_duplicates('Name')  # Only show one row per athlete

    return merged
```
